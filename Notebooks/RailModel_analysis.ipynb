```python
import pandas as pd
url='https://raw.githubusercontent.com/IvanKashitskii/RailModel/refs/heads/main/results/results_summary.csv'
df=pd.read_csv(url, sep=';',
encoding='utf-8')
df.head()
```

```python
df_clean=df.dropna()
cols=df_clean.columns[1:]
short=[
    "Единая без пер.",
    "Разделённая без пер.",
    "Единая с пер.",
    "Разделенная c пер."
       ]

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot([df_clean[c].values for c in cols], tick_labels=short, showmeans=True)
plt.title('Коэффициент порожнего пробега при разных режимах работы сети')
plt.ylabel('Коэффициент порожнего пробега')
plt.xticks(rotation=10, ha='right')
plt.ylim(0.23, 0.46)
plt.grid(True)
plt.tight_layout()
plt.show()
```
